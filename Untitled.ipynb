```python
import numpy as np
import os
ImageDataPath = "InstagramImages/"
temp = []
outputImage = np.zeros((128, 200, 3), dtype="uint8")
size = 100,128
for dir1 in os.listdir(ImageDataPath)[1:]:
    files = os.listdir(ImageDataPath + '/' + dir1)
    files.sort(key=lambda x: '{0:0>8}'.format(x).lower())
    pro_img = Image.open(os.path.join(ImageDataPath,dir1,'profile.jpg'))
    pro_img = pro_img.resize(size,Image.ANTIALIAS)
    pro_img = np.asarray(pro_img)
    if pro_img.shape != (128, 100, 3):
        pro_img = np.stack((pro_img,)*3, axis=-1)
    outputImage[0:128, 0:100] = pro_img
    for name in files[1:-1]:
        img = Image.open(os.path.join(ImageDataPath,dir1,name))
        img = img.resize(size,Image.ANTIALIAS)
        img = np.asarray(img)
        if img.shape != (128, 100, 3):
            img = np.stack((img,)*3, axis=-1)
        outputImage[0:128, 100:200] = img
        data = outputImage.flatten()
        temp.append(data)
```
